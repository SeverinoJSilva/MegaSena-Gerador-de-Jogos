```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
mega = pd.read_excel('Ultimos_resultado_mega_virada.xlsx')
mega
```

```python
plt.figure(figsize=(8,4))
plt.title('Frequencia MegaVirada', fontsize=12, pad=15)
plt.xlabel('Frequencia de dezenas sorteadas', fontsize=11)
plt.ylabel('Dezenas Sorteadas', fontsize=11)
mega.boxplot(column= ['b1','b2','b3','b4','b5','b6'])
#plt.savefig('Mega_freq_Virada(Boxplot).png', format='png', dpi = 300, bbox_inches='tight')   # Salvando o plot
```

```python
mega.describe()
```

```python
# Visualizao de eventos das dezenas por posio
plt.figure(figsize=(15,5))
plt.subplot(231) # 1 numero de linhas, 3 numero de colunas, 1 posio do grafico
plt.hist(x=mega.b1)
plt.subplot(232) # 1 numero de linhas, 3 numero de colunas, 2 posio do grafico
plt.hist(x=mega.b2)
plt.subplot(233) # 1 numero de linhas, 3 numero de colunas, 3 posio do grafico
plt.hist(x=mega.b3)
plt.subplot(234) # 1 numero de linhas, 3 numero de colunas, 4 posio do grafico
plt.hist(x=mega.b4)
plt.subplot(235) # 1 numero de linhas, 3 numero de colunas, 5 posio do grafico
plt.hist(x=mega.b5)
plt.subplot(236) # 1 numero de linhas, 3 numero de colunas, 6 posio do grafico
plt.hist(x=mega.b6)
plt.show()
```

```python
# Visualizao de eventos das dezenas por posio
plt.figure(figsize=(15,5))
plt.subplot(231) # 1 numero de linhas, 3 numero de colunas, 1 posio do grafico
plt.title('Bola1')
#plt.xticks(rotation=45)
sns.barplot(x = mega.b1.value_counts().index, y = mega.b1.value_counts().values)
plt.subplot(232) # 1 numero de linhas, 3 numero de colunas, 2 posio do grafico
plt.title('Bola2')
sns.barplot(x = mega.b2.value_counts().index, y = mega.b2.value_counts().values)
plt.subplot(233) # 1 numero de linhas, 3 numero de colunas, 3 posio do grafico
plt.title('Bola3')
sns.barplot(x = mega.b3.value_counts().index, y = mega.b3.value_counts().values)
plt.subplot(234) # 1 numero de linhas, 3 numero de colunas, 4 posio do grafico
plt.xlabel('Bola4')
sns.barplot(x = mega.b4.value_counts().index, y = mega.b4.value_counts().values)
plt.subplot(235) # 1 numero de linhas, 3 numero de colunas, 5 posio do grafico
plt.xlabel('Bola5')
sns.barplot(x = mega.b5.value_counts().index, y = mega.b5.value_counts().values)
plt.xlabel('Bola6')
plt.subplot(236) # 1 numero de linhas, 3 numero de colunas, 6 posio do grafico
plt.xlabel('Bola6')
sns.barplot(x = mega.b6.value_counts().index, y = mega.b6.value_counts().values)
plt.show()
```
